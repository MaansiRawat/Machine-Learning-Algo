# Decision Tree Classifier from Scikit-Learn on (IRIS) dataset:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

## 1. LOAD THE DATA :

```python
iris=datasets.load_iris()
```

## 2. Get the features & labels from the data: 

```python
x=iris.data
y=iris.target
```

```python
x
```

```python
y
```

## 3. Split the data into train and test data: 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

## 4. Standardize the data: (feature scaling)

```python
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

## 5. Choose Algo & set the hyperparameters: 

```python
tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
```

## 6. Train the Model:

```python
tree=tree.fit(x_train_std,y_train)
```

## 7. Make Prediction using the Model: 

```python
y_pred=tree.predict(x_test_std)
```

```python
y_pred
```

## pred & acc 

```python
print('Wrong Predictions out of Total:')
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()+(y_test!=y_pred).sum()))
```

```python
print('Percentage Accuracy:',100*accuracy_score(y_test,y_pred))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import tree 
clf=tree.DecisionTreeClassifier(criterion='entropy')
clf=clf.fit(iris.data,iris.target)
```

```python
import graphviz
dot_data=tree.export_graphviz(clf,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,rounded=True,special_characters=True)

graph=graphviz.Source(dot_data)
```
