## LOGISTIC REGRESSION (LR): 

C is [REGULARIZATION PARAMETER]

Logistic Regression used in binary classification & multiple but best in binary classification.



```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## LOAD THE DATA:

```python
iris=datasets.load_iris()
```

```python
iris
```

## Get the Features & labels from the Data:

```python
x=iris.data
y=iris.target
```

```python
x
```

```python
y
```

## Split into train and test:

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=0)
```

## Standardize the data: 

```python
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.fit_transform(x_test) 
```

```python
x_train_std
```

```python
x_test_std
```

## Choose Algo & set the hyperparameters: 

```python
#lr=LogisticRegression(C=1.0,random_state=0)   #C is Regulararization Parameter
lr=LogisticRegression(C=100,random_state=0)
```

## Train the Model: 

```python
lr.fit(x_train_std,y_train)
```

## Make Prediction using Model:

```python
y_pred=lr.predict(x_test_std)
```

```python
y_pred
```

```python
print("wrong Predictions out of total:")
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()+y_test!=y_pred).sum())
```

```python
print('Percentage Accuracy:',100*accuracy_score(y_test,y_pred))
```

## Confusion matrix: 

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
